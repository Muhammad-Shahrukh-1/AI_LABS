```python
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix ,accuracy_score
import pandas as pd
```

```python
df = pd.read_csv("dataset.csv")
df
```

```python
height = df["Height"]
weight = df["Weight"]
footsize = df["Foot_Size"]
gender = df["Gender"]
```

```python
features = list(zip(height,weight,footsize))
```

```python
model = DecisionTreeClassifier()
model = model.fit(features,gender)
```

```python
h = int(input("Enter height:: "))
w = int(input("Enter weight:: "))
f_s = int(input("Enter foot_size:: "))
new_entry = [[h,w,f_s]]
print("height = ",h)
print("weight = ",w)
print("foot_size = ",f_s)
predicted = model.predict(new_entry)
print(predicted)
```

```python
predicted2 = model.predict(features)
conf_mat = confusion_matrix(gender, predicted2)
print(conf_mat)
```

```python
accuracy = accuracy_score(predicted2,gender)
print(accuracy)
```

```python
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix ,accuracy_score
import pandas as pd
```

```python
df = pd.read_csv('random_dataset.csv')
print(df)
```

```python
le = preprocessing.LabelEncoder()
```

```python
age = df['Age']
salary = df['Salary']
experience = df['Experience']
education_level = le.fit_transform(df['Education Level'])
is_employed = df['Is Employed']
```

```python
features = list(zip(age,salary,experience,education_level))
label = is_employed
```

```python
model = DecisionTreeClassifier()
model = model.fit(features,label)
```

```python
predicted = model.predict(features)
print(predicted)
```

```python
conf_mat = confusion_matrix(label,predicted)
print(conf_mat)
```

```python
accuracy = accuracy_score(predicted,label)
accuracy
```
