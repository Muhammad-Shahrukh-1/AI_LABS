```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("homeprices.csv")
print(df)
```

```python
%matplotlib inline
plt.xlabel("area")
plt.ylabel("price")
print(plt.scatter(df.area,df.price,color="red",marker="+"))
```

```python
new_df = df.drop("price",axis="columns")
print(new_df)
```

```python
price = df.price
print(price)
```

```python
reg = LinearRegression()
reg.fit(new_df,price)
```

```python
x = 3300
predicted = reg.predict([[3300]])
print(predicted)
```

```python
m = reg.coef_
b = reg.intercept_
y = m*x+b
print(y)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("exam_scores_dataset.csv")
print(df)
```

```python
%matplotlib inline
plt.xlabel("Hours_Studied")
plt.ylabel("Exam_Score")
print(plt.scatter(df.Hours_Studied,df.Exam_Score,color="red",marker="+"))
```

```python
new_df = df.drop("Exam_Score",axis='columns')
print(new_df)
```

```python
exam_score = df.Exam_Score
print(exam_score)
```

```python
reg = LinearRegression()
reg.fit(new_df,exam_score)
```

```python
x = 40
predicted = reg.predict([[x]])
print(predicted)
```

```python
m = reg.coef_
b = reg.intercept_
y = m*x + b
print(y)
```
