```python
import heapq

# Function to implement max-priority queue insertion
def max_priority_queue(arr):
    # Create an empty list for the priority queue
    priority_queue = []
    
    for num in arr:
        # Insert each number as a negative value (to simulate max-heap)
        heapq.heappush(priority_queue, -num)
    
    # Convert the negative values back to positive and return the sorted list
    return [-heapq.heappop(priority_queue) for _ in range(len(priority_queue))]

# Input array
array = [4, 8, 1, 7, 3]

# Get the max-priority queue
result = max_priority_queue(array)
print("Max-Priority Queue:", result)
```

```python
class Vertex:
    def __init__(self, value):
        self.value = value
        self.neighbors = []  # List to store neighbors

    def add_neighbor(self, neighbor):
        """Add a neighbor to the vertex."""
        if neighbor not in self.neighbors:
            self.neighbors.append(neighbor)

    def get_neighbors(self):
        """Return the neighbors of the vertex."""
        return self.neighbors


class Graph:
    def __init__(self):
        self.vertices = {}  # Dictionary to store vertices with their values as keys

    def add_vertex(self, value):
        """Add a vertex to the graph."""
        if value not in self.vertices:
            self.vertices[value] = Vertex(value)

    def add_edge(self, from_vertex, to_vertex):
        """Add an edge between two vertices."""
        if from_vertex in self.vertices and to_vertex in self.vertices:
            self.vertices[from_vertex].add_neighbor(to_vertex)
            self.vertices[to_vertex].add_neighbor(from_vertex)  # Undirected graph

    def add_edges(self, edges):
        """Add multiple edges at once."""
        for from_vertex, to_vertex in edges:
            self.add_edge(from_vertex, to_vertex)

    def display_graph(self):
        """Display the graph as an adjacency list."""
        for vertex, vertex_obj in self.vertices.items():
            neighbors = ", ".join(vertex_obj.get_neighbors())
            print(f"{vertex}: {neighbors}")


# Example usage
# Create a graph
graph = Graph()

# Add vertices
graph.add_vertex("A")
graph.add_vertex("B")
graph.add_vertex("C")
graph.add_vertex("D")
graph.add_vertex("E")

# Add edges
edges = [("A", "B"), ("A", "C"), ("B", "D"), ("C", "D"), ("D", "E")]
graph.add_edges(edges)

# Display the graph
print("Graph as adjacency list:")
graph.display_graph()
```
